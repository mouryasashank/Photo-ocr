```
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize']=8,16
```

### en <-- English

```
reader = easyocr.Reader(['en'])
```

```
Image('sample.jpg')
```

```
output = reader.readtext('sample.jpg')
output
```

1st is the bounding box coordinates;
2nd is the text generated;
3rd is the confindence with text generated;

```
cord = output[-1][0]
```

```
x_min,y_min=[min(idx) for idx in zip(*cord)]
```

```
x_max,y_max=[max(idx) for idx in zip(*cord)]
```

```
img = cv2.imread('sample.jpg')
cv2.rectangle(img,(x_min,y_min),(x_max,y_max),(0,255,0),2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
